### Multiplication Table Generator

This program generates and displays the multiplication table for a given number. It prompts the user for input and handles basic validation to ensure a valid integer is entered.

```python
# Prompt the user for input
while True:
    try:
        num_str = input("Enter an integer to display its multiplication table: ")
        table_num = int(num_str)
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Display the multiplication table
print(f"\nMultiplication Table for {table_num}:")
for i in range(1, 11): # Display table up to 10
    result = table_num * i
    print(f"{table_num} x {i} = {result}")
```
