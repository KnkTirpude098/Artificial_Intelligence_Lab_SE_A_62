### Odd or Even Number Identifier

This program checks if a given integer is odd or even. It prompts the user to enter a number and provides a clear output, along with input validation.

```python
# Prompt the user for input
while True:
    try:
        num_str = input("Enter an integer to check if it's odd or even: ")
        num = int(num_str)
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Check if the number is odd or even
if num % 2 == 0:
    print(f"The number {num} is Even.")
else:
    print(f"The number {num} is Odd.")
```
