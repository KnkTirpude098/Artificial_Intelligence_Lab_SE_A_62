```python
# Simple Expert System: Medical Diagnosis

print("=== Simple Medical Expert System ===")

fever = input("Do you have a fever? (yes/no): ").lower()
cough = input("Do you have a cough? (yes/no): ").lower()
headache = input("Do you have a headache? (yes/no): ").lower()

print("\n--- Diagnosis Result ---")

if fever == "yes" and cough == "yes" and headache == "yes":
    print("You may have the Flu.")
elif fever == "yes" and cough == "yes":
    print("You may have a Common Cold.")
elif headache == "yes" and fever == "no":
    print("You may have a Migraine.")
elif fever == "no" and cough == "no" and headache == "no":
    print("You seem to be healthy.")
else:
    print("Symptoms are unclear. Please consult a doctor.")

print("\nThank you for using the Expert System.")
```
